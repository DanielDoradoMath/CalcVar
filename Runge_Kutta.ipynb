```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def runge_kutta (f, a, b, N, x0):
  dt = (b-a)/N
  t = np.linspace(a,b,N)
  x = [x0]

  for i in range(N-1):
    p = dt*f(t[i],x[i])
    q = dt*f(t[i]+dt*0.5,x[i]+p*0.5)
    r = dt*f(t[i]+dt*0.5,x[i]+q*0.5)
    s = dt*f(t[i]+dt,x[i]+r)
    x += [x[i]+(p+2*q+2*r+s)/6]
  
  return (t, np.array(x))
```

```python
def pend (t,x,g=9.81,l=1):
  theta = x[0]
  omega = x[1]

  d1 = omega
  d2 = -g*np.sin(theta)/l

  return (np.array([d1,d2]))
```

```python
time, vec = runge_kutta(pend, 0, 10, 1000, np.array([np.pi/6,0]))
```

```python
plt.figure()
plt.plot(time, vec[:,0])
plt.xlabel('Tiempo')
plt.ylabel('Ángulo')
plt.title
plt.show()
```

```python
def oscillator (t,x,k=2,m=1):
  pos = x[0]
  vel = x[1]

  d1 = vel
  d2 = -k*pos/m

  return (np.array([d1,d2]))
```

```python
time, vec = runge_kutta(oscillator, 0, 10, 1000, np.array([4,0]))
```

```python
plt.figure()
plt.plot(time, vec[:,0])
plt.xlabel('Tiempo')
plt.ylabel('Posición')

plt.show()
```
